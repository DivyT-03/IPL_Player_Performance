# Importing Data

"Training.csv" : File Containing various performance stats of a player in a match <br>
"Matches IPL 2020.csv" : Schedule of matches that occrued in IPL 2020<br>
"IPL 2020 Squads.csv" : Squads of each participating club/team in IPL 2020<br>
"Matches IPL 2008-2019.csv" : Schedule of matches that occrued in IPL 2020<br>

```python
import numpy as np
import pandas as pd
```

```python
Training = pd.read_csv("Training.csv")
Training.head()
```

```python
submission = pd.read_csv("sample_submission.csv")
submission.head()
```

```python
Matches_IPL_2008_2019 = pd.read_csv("Matches IPL 2008-2019.csv")
Matches_IPL_2008_2019.head()
```

```python
Matches_IPL_2020 = pd.read_csv("Matches IPL 2020.csv")
Matches_IPL_2020.head()
```

```python
Training[['Match_Id','Player']] = Training.Id.str.split("_",expand=True)
Training['Match_Id'] = Training['Match_Id'].astype('int64')
Training.head()
```

```python
Training = Training[['Match_Id','Player','Total Points']]
Training.head()
```

```python
submission[['Match_Id','Player']] = submission.Id.str.split("_",expand=True)
submission = submission[['Match_Id','Player','Total Points']]
submission['Match_Id'] = submission['Match_Id'].astype('int64')
submission.head()
```

```python
def  initial_getter(x):
    if x=="Kings XI Punjab":
        return "KXIP"
    if x=="Sunrisers Hyderabad":
        return "SRH"
        
    lst = x.split()
    initials = ""
    for i in lst:  # go through each name
        initials += i[0].upper()  # append the initial
    return initials
```

```python
Matches_IPL_2008_2019['team1'] = Matches_IPL_2008_2019['team1'].apply(lambda x: initial_getter(x))
Matches_IPL_2008_2019['team2'] = Matches_IPL_2008_2019['team2'].apply(lambda x: initial_getter(x))
Matches_IPL_2008_2019.head()
```

```python
Player_DB = Training[['Match_Id','Player','Total Points']].merge(Matches_IPL_2008_2019[['match_id','team1','team2']], right_on='match_id', left_on='Match_Id', how='inner')
Player_DB.drop('match_id',axis=1,inplace=True)
Player_DB.head()
```

```python
Submit_DB = submission[['Match_Id','Player','Total Points']].merge(Matches_IPL_2020[['match_id','team1','team2']], right_on='match_id', left_on='Match_Id', how='inner')
Submit_DB.drop('match_id',axis=1,inplace=True)
Submit_DB.head()
```

```python
DB = pd.concat([Player_DB,Submit_DB]) 
DB.head()
```

* Adding players average scores

```python
avg = DB.groupby('Player')['Total Points'].mean().reset_index()
avg.rename(columns = {'Total Points':'Avg'}, inplace = True) 
avg.head()
```

```python
DB1 = pd.merge(DB[['Player','Match_Id','team1','team2']], avg[['Player','Avg']], on = 'Player', how='inner')
DB1.head()
```

```python
Training_Set = pd.merge(DB[['Match_Id','Player']],DB1[['Player','Match_Id','team1','team2','Avg']], on = ['Match_Id','Player'], how='inner')
Training_Set.head()
```

```python
X = Training_Set[['Player','team1','team2','Avg']]
y = Training['Total Points']
```
